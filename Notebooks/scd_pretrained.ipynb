```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

def generate_results(audio_file_path, rttm_file_path):

  audio_file_name = os.path.splitext(os.path.basename(audio_file_path))[0]

  print("Loading model")
  pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization@2.1", use_auth_token='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')

  print("Predicting on audio")
  diarization = pipeline({'audio': audio_file_path})

  # load ground truth from RTTM file
  _, groundtruth = load_rttm(rttm_file_path).popitem()

  # Load audio
  y, sr = librosa.load(audio_file_path)
  t = np.arange(len(y)) / sr

  fig, ax = plt.subplots(figsize=(15, 6))

  # plot waveform
  ax.plot(t, y, color='grey', linewidth=0.5, label="Waveform")

  # determine range for vertical lines
  vline_range = (np.min(y) / 2, np.max(y) / 2)
  # prepare ground truth signal
  ground_truth_changes = [segment.start for segment in groundtruth.get_timeline()]

  # prepare prediction speaker change points
  predicted_changes = [segment.start for segment, _, _ in diarization.itertracks(yield_label=True)]

  # Create binary ground truth signal
  ground_truth_signal = np.zeros_like(y)
  for change in ground_truth_changes:
      index = int(change * sr)
      if index < len(y):
          ground_truth_signal[index] = 1

  # Create binary prediction signal
  predicted_signal = np.zeros_like(y)
  for change in predicted_changes:
      index = int(change * sr)
      if index < len(y):
          predicted_signal[index] = 1

  # Prepare annotation for ground truth
  groundtruth_anno = Annotation()
  for segment in groundtruth.get_timeline():
      groundtruth_anno[segment] = 'speaker'

  # Prepare annotation for predicted segments
  predicted_anno = Annotation()
  for segment, _, _ in diarization.itertracks(yield_label=True):
      predicted_anno[segment] = 'speaker'

  # Compute Diarization Error Rate
  der_metric = DiarizationErrorRate(collar=0.25, skip_overlap=False)  # collar is the amount of time tolerance for the diarization boundary
  der = der_metric(groundtruth_anno, predicted_anno)
  # print(f"Diarization Error Rate: {der}")
  print('ground_truth_signal: ', ground_truth_signal)
  print('predicted_signal', predicted_signal)
  # Compute AUC score
  auc_score = roc_auc_score(ground_truth_signal, predicted_signal)
  print(f"AUC Score: {auc_score}")
  # Compute precision, recall, and F1 score
  precision = precision_score(ground_truth_signal, predicted_signal)
  recall = recall_score(ground_truth_signal, predicted_signal)
  f1 = f1_score(ground_truth_signal, predicted_signal)

  print(f"Precision: {precision}")
  print(f"Recall: {recall}")
  print(f"F1 Score: {f1}")

  # Save all metrics in a dictionary
  metrics = {
        'DER': der,
        'AUROC': auc_score,
        'Precision': precision,
        'Recall': recall,
        'F1': f1
    }

  plot speaker change points for ground truth and prediction
  we use vertical lines, but with a smaller alpha (transparency level) and thicker line width
  for change in ground_truth_changes:
      ax.vlines(x=change, ymin=vline_range[0], ymax=vline_range[1], color='green', linewidth=2, alpha=0.7, label='Ground Truth' if change == ground_truth_changes[0] else "")

  for change in predicted_changes:
      ax.vlines(x=change, ymin=vline_range[0], ymax=vline_range[1], color='red', linewidth=2, alpha=0.7, linestyle='dashed', label='Predicted' if change == predicted_changes[0] else "")

  # set labels and title
  ax.set_xlabel('Time (s)')
  ax.set_ylabel('Amplitude')
  ax.set_title('Speaker Change Points')
  # ax.text(0.02, 0.02, 'ROC AUC Score: {:.2f}'.format(auc_score), transform=ax.transAxes)
  # add legend
  handles, labels = ax.get_legend_handles_labels()
  by_label = dict(zip(labels, handles))
  ax.legend(by_label.values(), by_label.keys())

  plt.tight_layout()
  output_dir = "/content/gdrive/MyDrive/test_data_trimmed/Outputs_pretrained"
  file_name = os.path.splitext(os.path.basename(audio_file_path))[0].replace('trimmed_', '').replace('.Mix-Headset','')
  plt.savefig(os.path.join(output_dir, f"{audio_file_name}.png"), bbox_inches="tight")
  plt.show()

  return metrics
```

```python
def pair_files(wav_files, rttm_files, dir_path):
    pairs = []
    for wav in wav_files:
        wav_stem = os.path.splitext(wav)[0].replace('trimmed_','').replace('.Mix-Headset','')
        for rttm in rttm_files:
            rttm_stem = os.path.splitext(rttm)[0].replace('trimmed_','')
            if wav_stem == rttm_stem:
                pairs.append((os.path.join(dir_path, wav), os.path.join(dir_path, rttm)))
    return pairs

def get_file_pairs(dir_path):
    # make sure dir_path is an absolute path
    dir_path = os.path.abspath(dir_path)

    # get list of all files in directory
    files = os.listdir(dir_path)
    # print(files)
    wav_files = [f for f in files if f.endswith('.wav')]
    rttm_files = [f for f in files if f.endswith('.rttm')]
    # get list of pairs of .wav and .rttm files with the same base name
    file_pairs = pair_files(wav_files, rttm_files, dir_path)
    # raise an error if no pairs are found
    if not file_pairs:
        raise ValueError(f"No paired files found in directory {dir_path}")
    return file_pairs

def get_processed_files(file_path='/content/gdrive/MyDrive/test_data_trimmed/Outputs_pretrained/processed_files_pretrained_new.txt'):
    try:
        with open(file_path, 'r') as file:
            return file.read().splitlines()
    except FileNotFoundError:
        return []

def add_processed_file(file_path, processed_file_path='/content/gdrive/MyDrive/test_data_trimmed/Outputs_pretrained/processed_files_pretrained_new.txt'):
    with open(processed_file_path, 'a') as file:
        file.write(file_path + '\n')

def convert_filename(name):
    return f'trimmed_{name}.Mix-Headset'
```

```python
import csv
dir_path = '/content/gdrive/MyDrive/test_data_trimmed'
# print(dir_path)
file_pairs = get_file_pairs(dir_path)

fieldnames = ['File_name', 'DER', 'AUROC', 'Recall', 'Precision', 'F1']

# # Read the list of already processed files
processed_files = get_processed_files()
print(processed_files)

# Write header row
with open('/content/gdrive/MyDrive/test_data_trimmed/Outputs_pretrained/new_results_pretrained_scd_2.csv', 'a', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

for pair in file_pairs:
  audio_file, rttm_file = pair
  # Skip if already processed
  if audio_file in processed_files:
      print(f'Skipping {audio_file}, already processed.')
      continue
  try:
    metrics = generate_results(audio_file, rttm_file)
    print(metrics)
  except Exception as e:
      print(f'Error generating results for file pair ({audio_file}, {rttm_file}): {e}')
      continue
  if not metrics:
      print(f'No results generated for file pair ({audio_file}, {rttm_file})')
      continue

  # Add the file name to the metrics dictionary
  metrics['File_name'] = os.path.splitext(os.path.basename(audio_file))[0].replace('trimmed_', '').replace('.Mix-Headset','')
  add_processed_file(audio_file)
  # print(f'Metrics for file {metrics['File_name']} saved in file')
  # Write the results to the csv file
  with open('/content/gdrive/MyDrive/test_data_trimmed/Outputs_pretrained/new_results_pretrained_scd_2.csv', 'a', newline='') as csvfile:
      writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
      writer.writerow(metrics)
```
